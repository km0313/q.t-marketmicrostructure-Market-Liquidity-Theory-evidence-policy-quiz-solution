- mktcap(market capitalization) and vp(trading volume) are mostly corrected

- this means mktcap and vp are not both need because of multicollinearity

```python
agf['log_mktcap']=np.log(agf['mktcap'])
agf['log_vp']=np.log(agf['vp'])
agf['turnover']=agf['vo']/agf['ibnosh']
agf['log_p']=np.log(agf['p'])
```

```python
import statsmodels.api as sm
def olsreg(x_col,y_col,agf=agf):
    X=agf[x_col].values
    X=sm.add_constant(X)
    result=sm.OLS(agf[y_col],X).fit()
    return result.summary()
```

```python

x_cols=['mktcap','vp','log_mktcap','log_vp','turnover',['turnover','log_p']]
y_col='bas100'
for x_col in x_cols:
    print(olsreg(x_col,y_col))
```

- 'mktcap','vp'는 너무 크기때문에 log를 취해야 설명력이 높아짐(R-squared가 올라감)
- 'log_vp'를 봤을때 trading volume(vp)가 1프로 증가하면 -0.8245xlog(1.01)만큼 bid-ask spread(종속변수)가 변한다고볼수있음
- 곱연산의 독립변수가 합연산으로 종속변수에 영향을줌
